### Bases de Datos en Python

Ya trabajamos con algunas estructuras de datos básicas en python (listas, tuplas, diccionarios), y también aprendimos a crear nuestras propias estructuras de datos (objetos). 

Existen un tipo de estructura de datos a la que estamos muy acostumbrados, que llamamos coloquialmente "base de datos". Estos objetos usualmente tienen disponibles, **para los usuarios de los datos**, las siguientes operaciones:

- Creación de nuevas columnas ("variables")
- Análisis de subconjuntos de datos (por ej, "filtros" en Excel)
- Tablas resúmen (por ej, "Tabla dinámica" en Excel -- "collapse" en Stata)
- Cruce de datos (por ej, "buscarv" o "buscarh" de Excel)
- Union de bases de datos ("append" de Stata)

Para lo anterior, existe una librería muy potente de Python llamada "Pandas".

Un elemento clave detrás de las bases de datos son los "indices". Es la forma que tienen las bases de datos de "buscar" la información de manera eficiente, permitiendo esto que las operaciones sobre la base de datos sea lo más rápido posible.

Busquemos la fecha de forma "ineficiente"

```python
def busca_fecha(fecha_a_buscar, lista_transacciones):
    '''
    Busca el número de transacciones para una fecha
    en particular
    '''
    
    for fecha, val in lista_transacciones:
        if fecha == fecha_a_buscar:
            return val

    return 0
```

```python
fecha_a_buscar = '2013-07-16'

busca_fecha(fecha_a_buscar, lista_transacciones)
```

```python
transacciones_por_fecha.loc[fecha_a_buscar]
```

```python
#ventas del un dia particular

%timeit busca_fecha(fecha_a_buscar, lista_transacciones)

%timeit transacciones_por_fecha.loc[fecha_a_buscar]

%timeit dicc_fechas[fecha_a_buscar]
```
